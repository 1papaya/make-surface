```python
import csv, os, rasterio, fiona
from scipy import stats
from rasterio import features
from shapely.geometry import Polygon
from fiona.crs import from_epsg
```

```python
def latlngToXY(ll):
    from math import pi, log, tan
    D2R = pi / 180.0
    A = 6378137.0
    x = (A * ll[0] * D2R)
    y = (A * log(tan((pi*0.25) + (0.5 * ll[1] * D2R))))
    return x,y

def classifyCrime(zArr,classes,weighting=1):
    'convert crime kernel surface into classed surface for polygonization'
    outRas = zeros(zArr.shape)
    zMax = zArr.max()
    zMin = zArr.min()
    zRange = zMax-zMin
    zInterval = zRange/float(classes)
    for i in range(0,classes):
        cClass = ((i*zInterval+zMin)*weighting+percentile(zArr, i/float(classes)*100))/(float(weighting)+1)
        outRas[where(zArr>cClass)] = cClass
    return outRas.astype(np.uint16)
```

```python
xys = []

sdir = '/Users/dnomadb/python-minis/data-doers/sfpd_incident_all_csv'
w = 1
for f in os.listdir(sdir):
    afile = os.path.join(sdir,f)
    year = int(f.replace('.csv','').split('_')[-1])
    with open(afile, 'r') as ofile:
        reader = csv.DictReader(ofile)
        crimelist = []
        bikelist = []
        for row in reader:
            crimelist.append(row['Category'])
            for i in row:
                bikeEval = row[i].lower()
                if bikeEval.find('bike') != -1 or bikeEval.find('bicycle') != -1:
                    bikelist.append(row)
                    for weight in range(w):
                        xys.append(latlngToXY([float(row['X']), float(row['Y'])]))
    w+=1
```
